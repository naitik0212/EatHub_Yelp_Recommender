```python
# restaurant_name_size = {}
one_restaurant_dict = {}
graph_dict = {}

graph_cuisne_array = []


# cuisine_names = list(df_champaign)
# cuisine_names = ["categories_mexican","categories_american","categories_italian","categories_japanese","categories_thai","categories_seafood","categories_chinese","categories_fast_food","categories_vegan","categories_vegetarian"]

cuisine_names = ["categories_mexican","categories_american","categories_vegan","categories_vegetarian","categories_seafood"]
graph_dict["name"] = "Cuisines"


for cuisine in cuisine_names:
    d = {}
    df_temp = df_champaign.loc[df_champaign[""+ cuisine + ""] == 1]
    df_temp = df_temp[["business_id","name","stars_y"]]
    one_restaurant_array = []
    one_cuisine_array = []
    one_cuisine_dict = {}
    
    if cuisine == "categories_mexican":
        actual_name = "Mexican"
    elif cuisine == "categories_american":
        actual_name = "American"
    elif cuisine == "categories_italian":
        actual_name = "Italian"
    elif cuisine == "categories_japanese":
        actual_name = "Japanese"
    elif cuisine == "categories_thai":
        actual_name = "Thai"
    elif cuisine == "categories_seafood":
        actual_name = "Seafood"
    elif cuisine == "categories_chinese":
        actual_name = "Chinese"
    elif cuisine == "categories_fast_food":
        actual_name = "Fast Food"
    elif cuisine == "categories_vegan":
        actual_name = "Vegan"
    elif cuisine == "categories_vegetarian":
        actual_name = "Vegetarian"  
    one_cuisine_dict["name"] = actual_name
    one_cuisine_dict["id"] = actual_name


    for index,row in df_temp.iterrows():    
        rest_info = {}
        rest_info["name"] = row["name"]
        temp_dict = {}
        temp_dict["name"] = row["name"]
        temp_dict["size"] = row["stars_y"]
        temp_dict["id"] = row["business_id"]
        rest_sub_info = [temp_dict]
        rest_info["children"] = rest_sub_info
        one_cuisine_array.append(rest_info)

    one_cuisine_dict["children"] = one_cuisine_array
    graph_cuisne_array.append(one_cuisine_dict)
    
graph_dict["children"] = graph_cuisne_array
import pprint
import json
r = json.dumps(graph_dict, indent=1)
pprint.pprint(graph_dict)
with open('Bubble-chart/bubble_tempe_data.json', 'w') as fp:
    json.dump(graph_dict, fp)

    
```
